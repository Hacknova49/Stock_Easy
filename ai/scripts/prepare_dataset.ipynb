```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw_dataset/BigBasket Products.csv")
df.head(5)
```

```python
print(df.columns)
print(df.shape)
```

```python
df = df[[
    "product",
    "category",
    "sale_price"
]]
```

```python
df.head(5)
```

```python
df.dropna(subset=["product", "category", "sale_price"], inplace=True)
```

```python
df["sale_price"] = pd.to_numeric(df["sale_price"], errors="coerce")
df.dropna(subset=["sale_price"], inplace=True)
```

```python
#dropping duplicates to prevetn ordering of the same items
df.drop_duplicates(subset=["product"], inplace=True)
```

```python
#added randome current stock
df["current_stock"] = np.random.randint(0, 100, size=len(df))
```

```python
#added randome avg_daily_sales
df["avg_daily_sales"] = np.random.randint(1, 10, size=len(df))
```

```python
df.head(3)
```

```python
df.describe()
```

```python
# Assume 20% retail margin
MARGIN = 0.20

df["supplier_cost"] = (df["sale_price"] * (1 - MARGIN)).round(2)
```

```python
#adding different supplier options
suppliers = ["SUP1", "SUP2", "SUP3"]
df["supplier_id"] = np.random.choice(suppliers, size=len(df))
```

```python
df.to_csv("../data/processed_dataset/inventory.csv", index=False)
```
